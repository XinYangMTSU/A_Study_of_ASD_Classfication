# DNN on ABIDE Database

# Experimentations

## I prepared data by using different atlases and connectivities, and save them into csv files
## Here is the naming rule: AtlasName_X_ConnetivityName, 
## For example 'ALL_X_corr.csv' means a text file that saves the feature vectors got by ALL atalas and Correlation connectivity
## For Atlas Power, please check https://drive.google.com/drive/folders/1lhc6csR7EWHYE1Bu0TjZlwU6rI7XuF88?usp=sharing
## For other Atlases and Y_data, please check https://drive.google.com/drive/folders/1hrGzfM2QH6ECqUWp23CRi0cfEtqMVGNm?usp=sharing
## I saved the data in two different places due to the limit of google drive storage, sorry for the inconvenience.
## If you want to prepare the data yourself, please check ABIDE_fMRI_Preprocessing.ipynb(Note that it may take hours to complete the preprocessing process).

```python
%load_ext autoreload
%autoreload 2
from ABIDE_fMRI_DNN import one_setting, get_names
from tensorflow.keras import optimizers
from tensorflow import keras
import numpy as np
Y_data = np.loadtxt('Data/Y.csv', delimiter=',')
np_seed = 200
tf_seed = 200
n_fold = 5
batchsz = 128

dropout_rate = 0.2
half = False
stratify = False
```

# AAL

```python
# Corrlation
optimizer = optimizers.Adam(lr=1e-4)
regularizer = keras.regularizers.L1L2(l1=0.02,l2=0.02)
X_data_file = 'Data/AAL_X_corr.csv'
X_data = np.loadtxt(X_data_file,delimiter=',')
title,save_name = get_names(X_data_file)
one_setting(np_seed, tf_seed, X_data, Y_data,
                           n_fold, batchsz, optimizer, regularizer, 
                           dropout_rate,half, stratify, title, save_name)
```

```python
# Part
optimizer = optimizers.Adam(lr=1e-4)
regularizer = keras.regularizers.L1L2(l1=0.02,l2=0.02)
X_data_file = 'Data/AAL_X_part.csv'
X_data = np.loadtxt(X_data_file,delimiter=',')
title,save_name = get_names(X_data_file)
one_setting(np_seed, tf_seed, X_data, Y_data,
                           n_fold, batchsz, optimizer, regularizer, 
                           dropout_rate,half, stratify, title, save_name)
```

```python
# Tangent
optimizer = optimizers.Adam(lr=1e-4)
regularizer = keras.regularizers.L1L2(l1=0.02,l2=0.02)
X_data_file = 'Data/AAL_X_tang.csv'
X_data = np.loadtxt(X_data_file,delimiter=',')
title,save_name = get_names(X_data_file)
one_setting(np_seed, tf_seed, X_data, Y_data,
                           n_fold, batchsz, optimizer, regularizer, 
                           dropout_rate,half, stratify, title, save_name)
```
